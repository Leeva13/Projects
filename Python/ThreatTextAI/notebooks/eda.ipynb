```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter
import nltk
from nltk.tokenize import word_tokenize

# Завантаження даних
df = pd.read_csv('../data/train.csv')

# 1. Розподіл класів
plt.figure(figsize=(8, 6))
sns.countplot(x='label', data=df)
plt.title('Розподіл класів')
plt.xlabel('Клас')
plt.ylabel('Кількість')
plt.show()

# 2. Аналіз довжини текстів
df['text_length'] = df['text'].apply(len)
plt.figure(figsize=(8, 6))
sns.histplot(df['text_length'], bins=20)
plt.title('Розподіл довжини текстів')
plt.xlabel('Довжина тексту')
plt.ylabel('Частота')
plt.show()

# 3. Найчастіші слова
nltk.download('punkt')
all_words = ' '.join(df['text']).lower()
tokens = word_tokenize(all_words)
word_freq = Counter(tokens).most_common(10)
print("Найчастіші слова:", word_freq)
```
